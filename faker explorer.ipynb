```python
from faker import Faker
```

```python
fake = Faker()
attrs = list(filter(lambda x: not x.startswith("_"), dir(fake)))
print(*attrs, sep="\n")
```

```python
for e in attrs:
    print(f"----------------- {e} --------------")
    help(eval(f"fake.{e}"))
    print("\n\n")
```

```python
class MyFaker:
    def __init__(self):
        self.i = 1
        
    def id(self):
        while True:
            yield self.i
            self.i += 1

fake = MyFaker()
for i in range(5):
    print(next(fake.id()))
```
